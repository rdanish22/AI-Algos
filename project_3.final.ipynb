````python
import copy
import math
import time
from heapq import heappush, heappop
from collections import deque
````

````python

import math

def distance(p1, p2):
    """
    Calculate the Euclidean distance between two points in 2D space.

    Parameters:
    p1 (tuple): Coordinates of the first point as (x1, y1).
    p2 (tuple): Coordinates of the second point as (x2, y2).

    Returns:
    float: The Euclidean distance between points p1 and p2.
    5.0
    """
    return math.sqrt((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2)
````

````python
from heapq import heappop, heappush

def find_path(world, start, goal):
    """
    A* search algorithm for pathfinding in a 2D grid.
    
    Parameters:
    - world (list of lists): Grid where '⬛' represents obstacles.
    - start (tuple): (row, col) of the starting position.
    - goal (tuple): (row, col) of the target position.

    Returns:
    - list: Shortest path [(row, col), ...] from start to goal or None if unreachable.

    Uses `g_score` for tracking movement cost and `came_from` to reconstruct the path.
    Prioritizes nodes with f-score = g-score + heuristic (distance function).
    """
    moves = [(-1, 0), (0, 1), (1, 0), (0, -1)]  # Possible movements: up, right, down, left
    open_set, came_from, g_score = [(0, start)], {start: None}, {start: 0}  # Init structures

    while open_set:
        _, current = heappop(open_set)  # Get node with lowest f-score
        if current == goal:  # Goal reached, reconstruct path
            path = []
            while current: path.append(current); current = came_from[current]
            return path[::-1]  # Reverse path to get correct order
        
        for move in moves:  # Explore neighbors
            neighbor = (current[0] + move[0], current[1] + move[1])
            if (0 <= neighbor[0] < len(world) and 0 <= neighbor[1] < len(world[0]) 
                and world[neighbor[0]][neighbor[1]] != '⬛'):  # Valid move check
                tentative_g = g_score[current] + 1  # Calculate new cost
                if neighbor not in g_score or tentative_g < g_score[neighbor]:  # Update if better
                    came_from[neighbor], g_score[neighbor] = current, tentative_g
                    heappush(open_set, (tentative_g + distance(neighbor, goal), neighbor))  # Add to queue
    return None  # No path found

````

````python
def update_butterflies(butterflies, robot_pos, world):
    """
    Moves butterflies using a combined attraction-repulsion force vector approach.

    Parameters:
    - butterflies (list of tuples): List of (x, y) positions.
    - robot_pos (tuple): (x, y) position of the robot.
    - world (list of lists): 2D grid for bounds.

    Returns:
    - list of tuples: Updated butterfly positions.
    """
    new_positions = []
    for butterfly in butterflies:
        # Initialize force vectors
        force_x, force_y = 0, 0
        
        # Attraction to the robot
        diff_x, diff_y = robot_pos[0] - butterfly[0], robot_pos[1] - butterfly[1]
        dist_robot = max(0.1, distance(butterfly, robot_pos))  # Avoid division by zero
        force_x += diff_x / dist_robot  # Normalize attraction force
        force_y += diff_y / dist_robot

        # Repulsion from other butterflies
        for other in butterflies:
            if other == butterfly:
                continue  # Don't repel yourself
            diff_x, diff_y = butterfly[0] - other[0], butterfly[1] - other[1]
            dist = max(0.1, distance(butterfly, other))
            if dist < 2:  # Only repel if within 2 units
                repulsion_strength = 1 / (dist * dist)
                force_x += diff_x * repulsion_strength
                force_y += diff_y * repulsion_strength

        # Normalize force vector (optional for smoother movement)
        total_force = math.sqrt(force_x**2 + force_y**2)
        if total_force > 0:
            force_x /= total_force
            force_y /= total_force

        # Compute new position with scaling
        move_scale = 0.8  # Adjust step size for smoother motion
        new_x = max(0, min(butterfly[0] + move_scale * force_x, len(world) - 1))
        new_y = max(0, min(butterfly[1] + move_scale * force_y, len(world[0]) - 1))
        new_positions.append((new_x, new_y))
    
    return new_positions

````

````python
def show_world(world, robot_pos, butterflies, goal):
    """
    Displays the current state of the world with the robot, butterflies, and goal.

    Parameters:
    - world (list of lists): A 2D grid representing the environment.
    - robot_pos (tuple): The (row, col) position of the robot, displayed as '🤖'.
    - butterflies (list of tuples): Positions of butterflies, displayed as '🦋'.
    - goal (tuple): The (row, col) position of the goal, displayed as '❤️'.

    The function creates a copy of the world, updates it with the robot, butterflies,
    and goal positions, then prints the updated world representation.
    """
    display = copy.deepcopy(world)  # Create a copy of the world
    
    # Place robot and goal
    display[robot_pos[0]][robot_pos[1]] = '🤖'
    display[goal[0]][goal[1]] = '❤️'
    
    # Place butterflies
    for b in butterflies:
        row, col = int(b[0]), int(b[1])  # Convert float positions to integers
        display[row][col] = '🦋'
    
    # Print the world
    for row in display:
        print(''.join(row))
    print()
````

````python
def simulation():
    """
    Runs the robot pathfinding simulation in a 21x21 grid world.

    The simulation includes:
    - A robot navigating from `start` (bottom-left) to `goal` (top-right) using A*.
    - Obstacles ('⬛') in a 3x3 block at the center.
    - Butterflies ('🦋') near the start, which move dynamically with attraction to the robot.

    The function:
    1. Creates the world grid and places obstacles.
    2. Finds the optimal path for the robot using `find_path()`.
    3. Iterates through the path, updating and displaying the world state.
    4. Updates butterfly positions using `update_butterflies()`.

    If no path is found, the function prints an error message.
    """
    # Create the world (empty grid with some obstacles)
    world = [['⬜'] * 21 for _ in range(21)]
    
    # Add obstacles (black squares in the center)
    for i in range(9, 12):
        for j in range(9, 12):
            world[i][j] = '⬛'
    
    # Set start and goal positions
    start, goal = (20, 0), (0, 20)  # Bottom-left to top-right
    
    # Place butterflies near the start
    butterflies = [(start[0], start[1]+i) for i in range(2, 10, 2)]
    
    # Find the robot's path
    path = find_path(world, start, goal)
    if not path:
        print("No path found!")
        return
    
    # Run the simulation: move robot along path, update butterflies, and display world
    for robot_pos in path:
        show_world(world, robot_pos, butterflies, goal)
        butterflies = update_butterflies(butterflies, robot_pos, world)


 
````

# Problem #2: Game Theory

We are going to conduct an experiment where two agents (players) employ the *Best Response with Inertia* algorithm against each other.

## Best Response with Inertia Algorithm

Here is the outline of the algorithm for each player:


```
On day zero, play a random action

On day t, do one of two things:

    * With high probability, play the best response to what your opponent did yesterday, or
    * With low probability, be lazy and just do what you did yesterday.
```

## Game

The particular 2x2 game being played is the Stag-Hunt game, with the matrix:


|  | | |
|---|---|---|
|  | S | H |
| S | 3, 3 | 0, 2 |
| H | 2, 0 | 2, 2 |


Player 1 (P1) is the `row` player; Player 2 (P2) is the `col` player.

````python
import random
import matplotlib.pyplot as plt
````

````python
# The actions available to each player. Don't modify. 
actions_P1 = ['S', 'H']
actions_P2 = ['S', 'H']

# Store the history. Don't modify. 
history_P1_action, history_P1_utility = [], []
history_P2_action, history_P2_utility = [], []
````

## get_utility

This function takes a joint action tuple and returns the utility for each player, packaged as a tuple. It is hardcoded for the Stag-Hunt game described above.

* **a** Tuple: joint action. a could be ('S', 'H'), for instance, which would signify that P1 has chosen to do 'S' and P2 has chosen to do 'H'.


**returns** 

* **utility** Tuple: the utility for each player as (utility of P1, utility of P2). For example, if the input is ('S', 'H'), the returned value is (0, 2) -- 0 for P1, 2 for P2. 

````python
def get_utility(a):
    if a[0] == 'S' and a[1] == 'S':
        return (3, 3)
    elif a[0] == 'S' and a[1] == 'H':
        return (0, 2)
    elif a[0] == 'H' and a[1] == 'S':
        return (2, 0)
    elif a[0] == 'H' and a[1] == 'H':
        return (2, 2)
````

````python
def best_response(opponent_last_action, player):
    """
    Determines the best response action for a player in the Stag-Hunt game
    based on the opponent's previous action.

    The Stag-Hunt game is a coordination game where players can either 
    cooperate (choose 'S' for Stag) or act independently (choose 'H' for Hare).
    The best strategy depends on the opponent’s last move:

    - If the opponent played 'S' (Stag), the best response is 'S' since it 
      yields the highest possible reward for both players.
    - If the opponent played 'H' (Hare), the best response is 'H' as it ensures 
      a safe, though suboptimal, payoff.

    Parameters:
    opponent_last_action (str): The last action played by the opponent ('S' or 'H').
    player (str): The player making the decision ('P1' or 'P2'). 
                  This parameter is currently unused but can be extended for future logic.

    Returns:
    str: The best response action ('S' or 'H').

    """
    return 'S' if opponent_last_action == 'S' else 'H'
````

````python

def run_experiment(num_trials, num_days, debug=True):
    if debug:
        print(f"Trial\tDay\tP1\tP2\n--------------------------")

    for trial in range(num_trials):
        if debug:
            print(f"--------------------------")

        # For each trial, pick random actions on day 0
        p1_last = random.choice(actions_P1)
        p2_last = random.choice(actions_P2)

        for day in range(num_days):

            # =============================
            # FIX: Implement Best Response with Inertia
            # =============================
            if day == 0:
                # Day 0: use the random actions chosen above
                action_P1 = p1_last
                action_P2 = p2_last
            else:
                # With probability 0.75, pick best response
                # With probability 0.25, do what you did yesterday (inertia)
                if random.random() < 0.75:
                    action_P1 = best_response(p2_last, player='P1')
                else:
                    action_P1 = p1_last

                if random.random() < 0.75:
                    action_P2 = best_response(p1_last, player='P2')
                else:
                    action_P2 = p2_last
            # =============================

            # Don't modify below:
            utility_P1, utility_P2 = get_utility((action_P1, action_P2))
            history_P1_action.append(action_P1)
            history_P1_utility.append(utility_P1)
            history_P2_action.append(action_P2)
            history_P2_utility.append(utility_P2)

            if debug:
                print(f"{trial}\t{day}\t{action_P1}\t{action_P2}")

            # Update "last" actions for next iteration
            p1_last = action_P1
            p2_last = action_P2

# Run the experiment with your chosen parameters (unchanged):
run_experiment(num_trials=20, num_days=10, debug=True)
````
