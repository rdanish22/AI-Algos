## Search


A general search pseudocode is as follows:

```
Here's a useful pseudocode for search, in general:

    Start with a frontier that contains the initial state
    Start with an empty explored set
    Repeat:
        If the frontier is empty, then no solution
        Remove a node from the frontier
        If node contains goal state, return the solution
        Add the node to the explored set
        Expand node, add resulting nodes to the frontier if they aren't already in the frontier or the explored set
```

Now, this is the general idea. How we deal with the frontier (implementation, queue, priority queue, stack, list, dictionary, and so on) and some of the conditions is going to differentiate whether this leads to breadth-first, depth-first, or A$^*$.


In class we talked about alphabetical order, but here we're dealing with locations in a list of lists. Let moves dictate the order. To arrive at the expected output, do not amend the order in which actions are arranged in moves (Up, Right, Down, Left).

````python
def bfs_helper(world, start, goal, moves):
    """
    Performs the core BFS search on a 2D grid world.
    It marks explored cells (except start and goal) with '🟨' in the world grid 
    and returns a dictionary mapping each cell to its predecessor if the goal is reached;
    otherwise, returns None.

    Args:
        world (List[List[str]]): 2D grid world.
        start (tuple): Starting coordinate (row, col).
        goal (tuple): Goal coordinate (row, col).
        moves (List[tuple]): Allowed moves (e.g., [(-1,0), (0,1), (1,0), (0,-1)]).

    Returns:
        dict or None: A mapping from cell to its predecessor if a path is found, otherwise None.
    """
    
    # Initialize BFS queue (frontier) with the start position
    frontier = [start]  # Using a list as a queue (FIFO behavior)
    
    # Dictionary to track the path: stores each cell's predecessor
    came_from = {start: None}
    
    # Set to track visited nodes to prevent re-exploration
    visited = {start}

    while frontier:
        # Dequeue the next node (FIFO for BFS)
        current = frontier.pop(0)

        # Mark explored cells with '🟨' (excluding start & goal)
        if current not in (start, goal):
            world[current[0]][current[1]] = '🟨'

        # If goal is reached, return the path mapping
        if current == goal:
            return came_from

        # Explore all possible moves (Up, Right, Down, Left)
        for dr, dc in moves:
            neighbor = (current[0] + dr, current[1] + dc)

            # Ensure neighbor is within bounds and not an obstacle or visited
            if (0 <= neighbor[0] < len(world) and 0 <= neighbor[1] < len(world[0])
                and world[neighbor[0]][neighbor[1]] != '⬛'
                and neighbor not in visited):

                # Mark as visited and enqueue for exploration
                visited.add(neighbor)
                frontier.append(neighbor)

                # Store the path by setting predecessor
                came_from[neighbor] = current

    # If no path is found, return None
    return None
````

````python
def breadth_first_search(world, start, goal, moves):
    """
    Performs a breadth-first search on a 2D grid world to find a path from start to goal.
    
    Visualizes the search process by marking explored cells with '🟨' and the final path
    with '🟩' in the world grid. The function prints the world state after the search
    is complete.

    Args:
        world (List[List[str]]): A 2D grid representing the search area where:
            - Empty cells are represented by '⬜'
            - Obstacles are represented by '⬛'
        start (tuple): Starting coordinate (row, col).
        goal (tuple): Goal coordinate (row, col).
        moves (List[tuple]): Allowed moves (e.g., [(-1,0), (0,1), (1,0), (0,-1)]).

    Returns:
        None: The function doesn't return a value but prints the visualization of the
        search process, showing:
        - Cells marked with '🟨' appeared in the frontier
        - Final path marked with '🟩'
        - Original world elements (obstacles, etc.) remain unchanged
    """
    # Work on a deep copy so the original world remains unchanged.
    world = deepcopy(world)
    came_from = bfs_helper(world, start, goal, moves)
    
    # Reconstruct the path from goal to start.
    path = []
    cell = goal
    while cell is not None:
        path.append(cell)
        cell = came_from[cell]
    path.reverse()
    # Mark the final path (except start and goal) with '🟩'.
    for cell in path:
        if cell not in (start, goal):
            world[cell[0]][cell[1]] = '🟩'
    for row in world:
         print("".join(row))
       
````

## Problem #2


Implement a depth-first search function to help the robot find a path to goal.

````python

def dfs_helper(grid, start, goal, moves):
    """
    Performs Depth-First Search (DFS) to explore a 2D grid and find a path from the start to the goal.

    DFS uses a stack-based approach to explore as deeply as possible before backtracking. This function 
    iterates over all possible moves in a reversed order to ensure a specific traversal pattern.

    The algorithm avoids obstacles ('⬛') and ensures each cell is visited only once to prevent infinite loops.
    It tracks the path using a dictionary (`came_from`) where each node points to its predecessor.

    Args:
        grid (List[List[str]]): A 2D grid representing the search space, where:
            - '⬜' represents empty walkable space.
            - '⬛' represents obstacles that cannot be traversed.
        start (tuple): The (row, col) coordinates of the starting position.
        goal (tuple): The (row, col) coordinates of the goal position.
        moves (List[tuple]): A list of allowed moves in the format (dr, dc), where:
            - (-1, 0) moves up
            - (0, 1) moves right
            - (1, 0) moves down
            - (0, -1) moves left
    
    Returns:
        dict or None:
            - If a path is found, returns a dictionary where each key is a cell, and its value is 
              the preceding cell in the path.
            - If no valid path exists, returns None.
    """

    # Stack for DFS (LIFO) behavior, initialized with the start position
    stack = [start]
    
    # Set to track visited cells and avoid re-exploration
    visited = set([start])
    
    # Dictionary to reconstruct the path: maps each cell to its predecessor
    came_from = {start: None}

    while stack:
        # Pop from the stack (LIFO) for depth-first exploration
        node = stack.pop()
        
        # If the goal is reached, return the path dictionary
        if node == goal:
            return came_from  

        # Explore all possible moves (Up, Right, Down, Left) in reversed order
        for move in reversed(moves):  
            next_pos = (node[0] + move[0], node[1] + move[1])
            
            # Check if the next position is within bounds, not an obstacle, and not visited
            if (0 <= next_pos[0] < len(grid) and 0 <= next_pos[1] < len(grid[0])
                and grid[next_pos[0]][next_pos[1]] != '⬛' and next_pos not in visited):
                
                # Mark as visited and add to stack for further exploration
                visited.add(next_pos)
                stack.append(next_pos)

                # Store the path by setting the predecessor
                came_from[next_pos] = node

    # If no path is found, return None
    return None  
````

````python
from copy import deepcopy

def depth_first_search(world, start, goal, moves):
    """
    Implements depth-first search (DFS) to find a path from start to goal in a 2D grid.

    DFS explores as deep as possible before backtracking, meaning it may not always find the shortest path 
    but will find a valid path if one exists. The function utilizes a helper function (dfs_helper) to 
    perform the main search and then reconstructs the path.

    Explored cells are marked with '🟨', and the final reconstructed path is marked with '🟩'.

    Args:
        world (List[List[str]]): 2D grid representing the search space, where:
            - '⬜' represents empty walkable space.
            - '⬛' represents obstacles that cannot be traversed.
        start (tuple): The (row, col) coordinates of the starting position.
        goal (tuple): The (row, col) coordinates of the goal position.
        moves (List[tuple]): A list of allowed moves in the format (dr, dc), where:
            - (-1, 0) moves up
            - (0, 1) moves right
            - (1, 0) moves down
            - (0, -1) moves left

    Returns:
        None: Prints the visual representation of the world using appropriate symbols:
            - '🟨' for explored cells (visited but not part of the final path).
            - '🟩' for the reconstructed path from start to goal.
            - '🤖' and '❤️' remain unchanged to indicate the start and goal positions.
    """

    # Create a deep copy of the world to avoid modifying the original grid
    grid = deepcopy(world)

    # Perform DFS using the helper function to find the path from start to goal
    came_from = dfs_helper(grid, start, goal, moves)

    # Mark all explored cells with '🟨' except for the start and goal
    for node in came_from:
        if node not in (start, goal):
            grid[node[0]][node[1]] = '🟨'

    # Reconstruct the final path from goal to start and mark it with '🟩'
    current = goal
    while current in came_from:
        previous = came_from[current]

        # Ensure start is not overwritten
        if previous and previous != start:
            grid[previous[0]][previous[1]] = '🟩'
        
        # Move backward along the path
        current = previous

    # Print the updated grid to visualize the DFS search and path
    for row in grid:
        print("".join(row))

````

## Problem #3


Implement an A$^*$ search function to help the robot find the optimal path to goal. For this problem, when calculating the accumulated path cost assume that every step the robot takes costs 1 unit. Use straight-line distance as your heuristic.


The output guidelines are similar to Problems 1 and 2. 

````python
def a_star_search(world, start, goal, moves):
    """
    Performs an A* search on a 2D grid world to find the optimal path from start to goal.

    A* uses a priority queue to explore paths efficiently by combining the cost-so-far (g-score)
    and an estimated remaining cost (heuristic). This implementation uses Euclidean distance 
    as the heuristic.

    Explored cells are marked with '🟨', and the final path is marked with '🟩'.

    Args:
        world (List[List[str]]): The 2D grid world, where:
            - '⬜' represents walkable space.
            - '⬛' represents obstacles.
        start (tuple): The (row, col) coordinates of the starting position.
        goal (tuple): The (row, col) coordinates of the goal position.
        moves (List[tuple]): A list of allowed moves, where:
            - (-1, 0) moves up
            - (0, 1) moves right
            - (1, 0) moves down
            - (0, -1) moves left

    Returns:
        None: Prints the visualization of the search process using appropriate symbols:
            - '🟨' for explored cells.
            - '🟩' for the reconstructed optimal path.
            - '🤖' and '❤️' remain unchanged to indicate the start and goal positions.
    """
    world, heuristic = deepcopy(world), lambda c: math.sqrt((c[0] - goal[0])**2 + (c[1] - goal[1])**2)
    open_list, came_from, g_score = [(heuristic(start), 0, start)], {start: None}, {start: 0}

    while open_list:
        _, current_g, current = heapq.heappop(open_list)
        if current == goal: break
        if current not in (start, goal): world[current[0]][current[1]] = '🟨'
        for dr, dc in moves:
            neighbor = (current[0] + dr, current[1] + dc)
            if 0 <= neighbor[0] < len(world) and 0 <= neighbor[1] < len(world[0]) and world[neighbor[0]][neighbor[1]] != '⬛':
                new_g = current_g + 1
                if neighbor not in g_score or new_g < g_score[neighbor]:
                    g_score[neighbor], came_from[neighbor] = new_g, current
                    heapq.heappush(open_list, (new_g + heuristic(neighbor), new_g, neighbor))

    cell, path = goal, []
    while cell: path.append(cell); cell = came_from.get(cell)
    for cell in reversed(path[:-1]): world[cell[0]][cell[1]] = '🟩'
    for row in world: print("".join(row))
````
