## Game

* We have two players: our original good robot and the new bad robot '👾'
* At each turn of the game, a player gets to move in the world
* Players only get to move according to `moves`
* '⬛' are obstacles that neither robot can traverse
* Each player knows the other player's current location
* Each player knows the location of the '❤️'
* '🤖' wants to get to the '❤️' without being zapped
* '👾' is trying to zap '🤖'
* '🤖' wins if it gets to the '❤️'
* '👾' wins if it zaps the good robot
* The game is over when either player wins or we've reached 82 plies (41 good robot moves and 41 bad robot moves)

## Adversarial Search


The MINIMAX pseudocode is as follows:

```
MINIMAX(state) returns action
    return argmax action in Actions MIN-VALUE(RESULT(state,action))
    

MAX-VALUE(state) returns value
if TERMINAL(state) return UTILITY(state)
v <- -inf
for each action in Actions(state)
    v <- MAX(v, MIN-VALUE(RESULT(state, action))
return v


MIN-VALUE(state) returns value
if TERMINAL(state) return UTILITY(state)
v <- +inf
for each action in Actions(state)
    v <- MIN(v, MAX-VALUE(RESULT(state, action))
return v
```


## Notes. 

* MINIMAX is used to model the opponent as a rational player that will also try it's best to win. When moves are simulated in the tree, that's just what they are, simulations. The actual board only changes once MINIMAX returns an action and that action is taken.

* Use a depth parameter to control the number of times the algorithm ping pongs between the calls to the min_value and max_value functions. A depth of 2 would mean P1_MAX's (caller of MINIMAX) moves are simulated and then P2_MIN's moves are simulated, and that's it.

````python
def is_valid(pos, world):
    """
    Checks if a given position is within the bounds of the grid and not an obstacle.

    Args:
        pos (Tuple[int, int]): The position (row, col) to check.
        world (List[List[str]]): The 2D grid representing the environment.

    Returns:
        bool: True if the position is within bounds and not an obstacle ('⬛'), False otherwise.
    """
    r, c = pos  # Extract row and column indices.
    return 0 <= r < len(world) and 0 <= c < len(world[0]) and world[r][c] != '⬛'
````

````python
def get_valid_moves(pos, world, moves):
    """
    Return a list of valid positions from the given position using the move offsets.

    Args:
        pos (Tuple[int, int]): The current position (row, col) of the robot.
        world (List[List[str]]): The 2D grid representing the environment.
        moves (List[Tuple[int, int]]): List of possible moves as (row_offset, col_offset).

    Returns:
        List[Tuple[int, int]]: A list of valid positions where the robot can move.
    """
    valid = []  # List to store valid moves

    for dr, dc in moves:  # Iterate over all possible move offsets
        new_pos = (pos[0] + dr, pos[1] + dc)  # Compute the new position

        # Check if the new position is within bounds and is not an obstacle
        if is_valid(new_pos, world):
            valid.append(new_pos)  # Add to the list of valid moves

    return valid  # Return the list of valid moves
````

````python
import math

def euclidean(position1, position2):
    """
    Computes the Euclidean distance between two positions in a 2D grid.

    Args:
        position1 (Tuple[int, int]): The first position (row, col).
        position2 (Tuple[int, int]): The second position (row, col).

    Returns:
        float: The Euclidean distance between pos1 and pos2.
    """
    return math.sqrt((position1[0] - position2[0])**2 + (position1[1] - position2[1])**2)
````

````python
def terminal_state(good, bad, goal):
    """
    Determines if the game has reached a terminal state.

    The game ends if:
    - The good robot (🤖) has reached the goal.
    - The good robot is adjacent (including diagonals) to the bad robot (👾), meaning it got zapped.

    Args:
        good (Tuple[int, int]): The current position of the good robot (row, col).
        bad (Tuple[int, int]): The current position of the bad robot (row, col).
        goal (Tuple[int, int]): The target goal position (row, col).

    Returns:
        bool: True if the game has ended, False otherwise.
    """
    # Check if the good robot reached the goal
    if good == goal:
        return True

    # Check if the good robot is adjacent to the bad robot (zapped)
    if abs(good[0] - bad[0]) <= 1 and abs(good[1] - bad[1]) <= 1:
        return True

    return False  # Game is still ongoing

````

````python
import math

def max_value(agent, opponent, target, depth, world, agent_is_good, mode):
    """
    Computes the maximum utility value and best move for the maximizing player in the minimax algorithm.

    Args:
        agent (Tuple[int, int]): The current position of the maximizing agent.
        opponent (Tuple[int, int]): The opponent's position.
        target (Tuple[int, int]): The goal position.
        depth (int): The remaining search depth.
        world (List[List[str]]): The 2D grid representing the environment.
        agent_is_good (bool): True if evaluating for the good robot, False for the bad robot.
        mode (int): Determines the utility function for the bad robot.

    Returns:
        Tuple[float, Tuple[int, int]]: The maximum utility value and the best move.
    """
    # Base case: If depth is zero, compute utility.
    if depth == 0:
        if agent_is_good:
            # Good robot's utility: Distance from bad - Distance from goal
            return (float('-inf') if euclidean(agent, opponent) <= math.sqrt(2)
                    else euclidean(opponent, agent) - euclidean(agent, target)), agent
        else:
            # Bad robot's utility based on the mode
            dist_to_agent = euclidean(agent, opponent)
            dist_to_goal = euclidean(agent, target)
            return (-dist_to_agent if mode == 2 else -(dist_to_goal + dist_to_agent)), agent

    best_score, best_step = float('-inf'), None  # Initialize best evaluation and move.

    # Evaluate all possible moves and update the best one.
    for offset in moves:
        next_pos = (agent[0] + offset[0], agent[1] + offset[1])
        if is_valid(next_pos, world):  # Ensure the new position is within valid bounds.
            score, _ = min_value(next_pos, opponent, target, depth - 1, world, agent_is_good, mode)
            if score > best_score:  # Update if a better move is found.
                best_score, best_step = score, next_pos

    return best_score, best_step


def min_value(agent, opponent, target, depth, world, agent_is_good, mode):
    """
    Computes the minimum utility value and best move for the minimizing player in the minimax algorithm.

    Args:
        agent (Tuple[int, int]): The current position of the minimizing agent.
        opponent (Tuple[int, int]): The opponent's position.
        target (Tuple[int, int]): The goal position.
        depth (int): The remaining search depth.
        world (List[List[str]]): The 2D grid representing the environment.
        agent_is_good (bool): True if evaluating for the good robot, False for the bad robot.
        mode (int): Determines the utility function for the bad robot.

    Returns:
        Tuple[float, Tuple[int, int]]: The minimum utility value and the best move.
    """
    # Base case: If depth is zero, compute utility.
    if depth == 0:
        if agent_is_good:
            # Good robot's utility: Distance from bad - Distance from goal
            return (float('-inf') if euclidean(agent, opponent) <= math.sqrt(2)
                    else euclidean(opponent, agent) - euclidean(agent, target)), agent
        else:
            # Bad robot's utility based on the mode
            dist_to_agent = euclidean(agent, opponent)
            dist_to_goal = euclidean(agent, target)
            return (-dist_to_agent if mode == 2 else -(dist_to_goal + dist_to_agent)), agent

    worst_score, worst_step = float('inf'), None  # Initialize worst-case evaluation and move.

    # Evaluate all possible moves and update the worst one.
    for offset in moves:
        next_pos = (opponent[0] + offset[0], opponent[1] + offset[1])
        if is_valid(next_pos, world):  # Ensure the new position is within valid bounds.
            score, _ = max_value(agent, next_pos, target, depth - 1, world, agent_is_good, mode)
            if score < worst_score:  # Update if a worse move is found.
                worst_score, worst_step = score, next_pos

    return worst_score, worst_step
````

````python
def minimax(agent, opponent, target, depth, is_maximizing, world, agent_is_good, mode):
    """
    Minimax algorithm wrapper that selects the best move for the current agent.

    Depending on whether the agent is maximizing or minimizing, this function calls
    either `max_value` (for maximizing player) or `min_value` (for minimizing player)
    to compute the optimal move.

    Args:
        agent (Tuple[int, int]): The current position of the agent making the move.
        opponent (Tuple[int, int]): The opponent's position.
        target (Tuple[int, int]): The goal position.
        depth (int): The remaining search depth.
        is_maximizing (bool): True if the current agent is maximizing, False if minimizing.
        world (List[List[str]]): The 2D grid representing the environment.
        agent_is_good (bool): True if evaluating for the good robot, False for the bad robot.
        mode (int): Determines the utility function for the bad robot.

    Returns:
        Tuple[float, Tuple[int, int]]: The computed utility value and the best move.
    """
    # If the agent is the maximizing player, call max_value to find the best possible move.
    # Otherwise, call min_value to find the move that minimizes the opponent's advantage.
    return (
        max_value(agent, opponent, target, depth, world, agent_is_good, mode)
        if is_maximizing 
        else min_value(agent, opponent, target, depth, world, agent_is_good, mode)
    )
````

````python

import math
import random
import copy


def run_robot_game(world, good_start, bad_start, goal, moves, good_strategy, bad_strategy, question):
    """
    Runs a turn-based game where Good (🤖) and Bad (👾) robots move.
    Good uses minimax (depth 2) or random moves; Bad uses minimax (depth 4 for "minimax_4", else 2) or random moves.
    The game stops if Good reaches the goal, is adjacent to Bad, or after 41 rounds.
    The final board shows Good's path as '🟩', Bad's path as '🟥', and announces the winner.
    """
    board, g, b, gpath, bpath = copy.deepcopy(world), good_start, bad_start, [], []  # Initialize board and positions.
    for _ in range(41):  # Up to 41 rounds.
        if terminal_state(g, b, goal): break  # End if terminal state reached.
        g = (minimax(g, b, goal, 2, True, world, True, question)[1]
             if good_strategy=="minimax" else (random.choice(get_valid_moves(g, world, moves)) if get_valid_moves(g, world, moves) else g))
        gpath.append(g)
        if terminal_state(g, b, goal): break
        d = 4 if bad_strategy=="minimax_4" else 2  # Set Bad's minimax depth.
        b = (minimax(b, g, goal, d, True, world, False, question)[1]
             if bad_strategy.startswith("minimax") else (random.choice(get_valid_moves(b, world, moves)) if get_valid_moves(b, world, moves) else b))
        bpath.append(b)
        if terminal_state(g, b, goal): break
    # Mark paths: '🟩' for Good (if cell is blank) and '🟥' for Bad (if cell is blank or already Good's mark).
    for pos, char in [(p, '🟩') for p in gpath] + [(p, '🟥') for p in bpath]:
        if (char=='🟩' and board[pos[0]][pos[1]]=='⬜') or (char=='🟩' and board[pos[0]][pos[1]]=='⬜') or (char=='🟥' and board[pos[0]][pos[1]] in ['⬜','🟩']):
            board[pos[0]][pos[1]] = char
    for row in board: print("".join(row))  # Print final board.


````
